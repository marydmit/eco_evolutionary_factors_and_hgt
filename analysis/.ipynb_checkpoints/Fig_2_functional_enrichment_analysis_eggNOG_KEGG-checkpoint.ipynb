```python
# Data manipulation
import numpy as np
import pandas as pd

# Statistics
from scipy.stats import binom_test
from statsmodels.stats.multitest import multipletests

# Visualization
import seaborn as sns
from matplotlib import pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages
plt.rcParams['svg.fonttype'] = 'none'
```

```python
# Importing file containing gene pairs with detected transfers along with support value information, gene and species distances. 
pairs_w_transfers = pd.read_csv("/Users/mdmit/Documents/hgt_project/file_repository/gene_pairs_with_transfers_score_80_and_info_on_support_dist_and_seq_identity.tsv", sep = "\t")
print(pairs_w_transfers.shape)

# Filtering out problematic genomes and specI clusters.
problematic_genomes = ["1932669.SAMEA43981918", "550.SAMEA3856670", "1916230.SAMN05904579", "1618951.SAMN03319549",
                       "1802238.SAMN04314631", "74031.SAMN03943304", "663.SAMN04088202"]
pairs_w_transfers = pairs_w_transfers.loc[~(pairs_w_transfers["genome_1"].isin(problematic_genomes)) & ~(pairs_w_transfers["genome_2"].isin(problematic_genomes)), :]
problematic_clusters = ["Cluster22", "Cluster2375", "Cluster4851", "Cluster5010", "Cluster5514"]
pairs_w_transfers = pairs_w_transfers.loc[~(pairs_w_transfers["speci_1"].isin(problematic_clusters)) & ~(pairs_w_transfers["speci_2"].isin(problematic_clusters)), :]
print(pairs_w_transfers.shape)
```

```python
# Retaining only transfer nodes where at least one transferred gene pair has branch support values above 0.5.
node_ids_good_support = pairs_w_transfers.loc[pairs_w_transfers["min_support"] >= 0.5, ["family_id", "node_id"]].drop_duplicates()
print(node_ids_good_support.shape)

transfers_good_support = pairs_w_transfers.loc[(pairs_w_transfers["family_id"] + pairs_w_transfers["node_id"]).isin(node_ids_good_support["family_id"] + node_ids_good_support["node_id"]), :]
print(transfers_good_support.shape)
```

```python
# For each transfer node, calculating the average sequence identity and species distance.
# For the gene distance and sequence identity to be accurrate, we will only take pairs with at least 50% overlap.
transfers_good_support.loc[transfers_good_support["seq_coverage"] < 50, "gene_distance"] = np.nan  # By default, pandas ignores nan values when calculating the mean.
transfers_good_support.loc[transfers_good_support["seq_coverage"] < 50, "seq_identity"] = np.nan  # By default, pandas ignores nan values when calculating the mean.
transfer_node_distances = transfers_good_support.groupby(["family_id", "node_id"]).agg({'distance': 'mean', 'gene_distance': 'mean', 'seq_identity': 'mean'}).reset_index()
transfer_node_distances
```

```python
# Visualising transfer node average species and gene distance distributions.
sns.set(font_scale = 2)
sns.set_style("ticks")
figures, axes = plt.subplots(1, 1, figsize = (10, 10))

cbar_ax = figures.add_axes([0.25, -0.01, 0.52, .025])

min_distance = 0
max_distance = 2
min_gene_distance = 0
max_gene_distance = 1
bin_number = 80

histogram_w_transfers  = np.histogram2d(x = transfer_node_distances["distance"],
                                        y = transfer_node_distances["gene_distance"],
                                        bins = bin_number,
                                        range = [[min_distance, max_distance], [min_gene_distance, max_gene_distance]])
histogram_w_transfers[0][histogram_w_transfers[0] == 0] = np.nan
ratio = np.log10(histogram_w_transfers[0])
ratio = np.transpose(ratio)


sns.heatmap(ratio, annot = False, cmap = "BrBG", ax = axes, center = 0, vmax = 6.3, cbar_ax = cbar_ax,
            cbar_kws={'label': r'$log_{10}( Transfer\ Nodes )$', 'orientation': 'horizontal'},
            xticklabels = True, yticklabels = True)


axes.invert_yaxis()
axes.set_yticks([0, 0.25*bin_number, 0.5*bin_number, 0.75*bin_number, bin_number])
axes.set_ylabel("Gene Tree Distance")
axes.set_yticklabels(["{:.2f}".format(x) for x in [0, 0.25*max_gene_distance, 0.5*max_gene_distance, 0.75*max_gene_distance, max_gene_distance]], rotation = 0)

axes.set_xticks([0, 0.25*bin_number, 0.5*bin_number, 0.75*bin_number, bin_number])
axes.set_xlabel("Species Tree Distance")
axes.set_xticklabels(["{:.1f}".format(x) for x in [0, 0.25*max_distance, 0.5*max_distance, 0.75*max_distance, max_distance]], rotation = 0)
axes.set_facecolor("#bfbfbf")

for _, spine in axes.spines.items():
    spine.set_visible(True)
```

# Functional Enrichment Analysis

```python
def group_terms_into_categories(families_annotated_w_terms, term_column, grouping_vector):
    families_annotated_w_terms["category"] = families_annotated_w_terms[term_column].map(grouping_vector)
    families_annotated_w_terms = families_annotated_w_terms.explode("category")
    families_annotated_w_terms = families_annotated_w_terms.drop_duplicates(["family_id", "category"])
    families_annotated_w_terms = families_annotated_w_terms.loc[:, ["family_id", "category", "number_genes"]]
    return families_annotated_w_terms

def count_transfer_nodes_per_bin(transfer_nodes_annotated, species_distance_breaks, gene_distance_breaks):
    
    all_bins_count_table = pd.DataFrame({"category": [], "counts_in_bin": [], "total_counts": [],  "fraction_bin": [], "species_bin": [], "gene_bin": []})
    for i in range(len(species_distance_breaks)):

        selected_transfer_nodes = transfer_nodes_annotated.loc[(transfer_nodes_annotated["distance"] >= species_distance_breaks[i][0]) & (transfer_nodes_annotated["distance"] < species_distance_breaks[i][1]), :]

        for j in range(len(gene_distance_breaks)):

            selected_transfer_nodes_to_count = selected_transfer_nodes.loc[(selected_transfer_nodes["gene_distance"] >= gene_distance_breaks[j][0]) & (selected_transfer_nodes["gene_distance"] < gene_distance_breaks[j][1]), :]
            counts_categories = selected_transfer_nodes_to_count.groupby("category").node_id.count().reset_index()
            counts_categories.columns = ["category", "counts_in_bin"]
            counts_categories["total_counts"] = len(selected_transfer_nodes_to_count.drop_duplicates("node_id"))
            counts_categories["fraction_bin"] = counts_categories["counts_in_bin"] / counts_categories["total_counts"]
            counts_categories["species_bin"] = species_distance_breaks[i][1]
            counts_categories["gene_bin"] = gene_distance_breaks[j][1]
            all_bins_count_table = pd.concat([all_bins_count_table, counts_categories])
    
    return all_bins_count_table
    
def perform_enrichment_test(count_table_per_bin, background_estimate):

    count_table_per_bin_w_bg = count_table_per_bin.merge(background_estimate)
    count_table_per_bin_w_bg["odds_ratio"] = np.log2(count_table_per_bin_w_bg["fraction_bin"] / count_table_per_bin_w_bg["fraction_total"])
    count_table_per_bin_w_bg["p_value"] = count_table_per_bin_w_bg.apply(lambda x: binom_test(x["counts_in_bin"], x["total_counts"], x["fraction_total"]), axis = 1)
    count_table_per_bin_w_bg["p_adjust"] = multipletests(count_table_per_bin_w_bg["p_value"].values)[1]
    
    return count_table_per_bin_w_bg

def convert_pvalue_to_asterisks(pvalue):
    if pvalue <= 0.001:
        return r'${\ast}{\ast}{\ast}$'
    elif pvalue <= 0.01:
        return r'${\ast}{\ast}$'
    if pvalue <= 0.05:
        return r'$\ast$'
    else:
        return "ns"
```

```python
# Breaking the transfer node landscape into three equal parts (+ looking at most recent transfers separately).
species_distance_breaks = [(0.08, 0.72), (0.72, 1.36), (1.36, 2.0)]
gene_distance_breaks = [(0, 0.01 + 1E-15), (0, 0.05 + 1E-15), (0, 0.25), (0.25, 0.5), (0.5, 0.75)]  # lower distance bound is included, higher is not
```

## COGs from eggNOG

```python
eggnog_categories = ["J", "K", "L",
                     "D", "M", "N", "O", "T", "U", "V",
                     "C", "E", "F", "G", "H", "I", "P", "Q",
                     "S"]
# We remove categories A, B, W and Z since there's too few observations in these categories.
cat_type = pd.CategoricalDtype(categories = eggnog_categories, ordered = True)
```

```python
# Uploading table containing functional mappings for each gene family.
families_annotated_with_eggnog = pd.read_csv("functional_enrichment_all_families_with_eggnog_annotations.csv")
transfer_nodes_annotated_w_eggnog = pd.merge(transfer_node_distances, families_annotated_with_eggnog, on = "family_id")
```

```python
# Plotting background distribution of COGs.
cog_counts_all_families = families_annotated_with_eggnog.groupby("category").number_genes.sum().reset_index()
cog_counts_all_families["fraction_total"] = cog_counts_all_families["number_genes"] / sum(families_annotated_with_eggnog.drop_duplicates(["family_id"])["number_genes"])
cog_counts_all_families["category"] = cog_counts_all_families["category"].astype(cat_type)
cog_counts_all_families

sns.set(font_scale = 1.5, style = "ticks")
figure, axes = plt.subplots(1, 1, figsize = (10, 5))
plot_palette = sns.color_palette("Purples", 3).as_hex() + sns.color_palette("Greens", 7).as_hex() + sns.color_palette("Oranges", 8).as_hex() + ["#DDDDDD"]
sns.barplot(data = cog_counts_all_families, x = "category", y = "fraction_total", ax = axes, palette = plot_palette)
axes.set_title("Background Expectation")
axes.set_xlabel("COG Category")
axes.set_ylabel("Fraction of All Genes")
```

```python
# Counting the number of transfer nodes for each category in each species and gene distance bin.
cog_counts_per_bin = count_transfer_nodes_per_bin(transfer_nodes_annotated_w_eggnog, species_distance_breaks, gene_distance_breaks)

# Now that we have counts per bin and background counts, we can perform functional enrichment analysis.
cog_enrichment_per_bin = perform_enrichment_test(cog_counts_per_bin, cog_counts_all_families)
cog_enrichment_per_bin.to_csv("results_functional_enrichment_analysis_COG_categories.csv", index = False)
```

```python
# Counting the number of transfer nodes for each antimicrobial class in each species and gene distance bin.
kegg_args_counts_per_bin = count_transfer_nodes_per_bin(transfer_nodes_annotated_w_kegg_args, species_distance_breaks, gene_distance_breaks)
kegg_args_counts_per_bin["category"] = kegg_args_counts_per_bin["category"].apply(lambda x: x.split()[0]).astype(kegg_args_cat_type)

# Now that we have counts per bin and background counts, we can perform functional enrichment analysis.
kegg_args_enrichment_per_bin = perform_enrichment_test(kegg_args_counts_per_bin, kegg_args_counts_all_families)
kegg_args_enrichment_per_bin.to_csv("results_functional_enrichment_analysis_KEGG_ARGs.csv", index = False)
```

```python
# Changing species and gene bin format to string.
kegg_args_enrichment_per_bin["species_bin_str"] = kegg_args_enrichment_per_bin["species_bin"].apply(lambda x: "{:.2f}-{:.2f}".format(x - 0.64, x)).astype("category")
kegg_args_enrichment_per_bin["gene_bin_str"] = kegg_args_enrichment_per_bin["gene_bin"].apply(lambda x: "{:.2f}-{:.2f}".format(max(x - 0.25, 0), x)).astype("category")

# For significance annotation, we convert p-value labels to stars.
kegg_args_enrichment_per_bin["significance"] = kegg_args_enrichment_per_bin["p_adjust"].apply(convert_pvalue_to_asterisks)

# Plotting all the categories except selected ones.
figures, axes = plt.subplots(3, 5, figsize = (22, 22), sharex = True, sharey = True)
sns.set(font_scale = 2)
sns.set_style("ticks")
cbar_ax = figures.add_axes([0.32, 0.02, 0.36, 0.025])

for i in range(len(kegg_args_categories)):
    category = kegg_args_categories[i]
    row = i // 5
    column = i % 5
    
    selection_category = kegg_args_enrichment_per_bin.loc[kegg_args_enrichment_per_bin["category"] == category, :]    
    significance = pd.pivot(selection_category, index = "gene_bin_str", columns = "species_bin_str", values = "significance")
    odds_ratios  = pd.pivot(selection_category, index = "gene_bin_str", columns = "species_bin_str", values = "odds_ratio")

    if i == 0:
        sns.heatmap(data = odds_ratios, cmap = "RdBu", square = True, annot = significance, fmt = '', ax = axes[row, column],
                    vmin = -4.05, vmax = 6.3, center = 0, linecolor = "black", linewidths = 1, annot_kws={"size": 18},
                    cbar_ax = cbar_ax, cbar_kws={'label': r'$log_2\left(\frac{Transfer\ Events\ Observed}{Transfer\ Events\ Expected}\right)$', 'orientation': 'horizontal'})
    else:
        sns.heatmap(data = odds_ratios, cmap = "RdBu", square = True, annot = significance, fmt = '', ax = axes[row, column],
                    vmin = -4.05, vmax = 6.3, center = 0, cbar = False, linecolor = "black", linewidths = 1, annot_kws={"size": 18})
        
    axes[row, column].set_xlabel("")
    axes[row, column].tick_params('y', rotation = 0)

    axes[row, column].set_ylabel("")
    axes[row, column].invert_yaxis()
    axes[row, column].set_xticklabels(axes[row, column].get_xticklabels(), rotation = -30, ha = "left", rotation_mode = "anchor")
    
    axes[row, column].set_title(category)

    for _, spine in axes[row, column].spines.items():
        spine.set_visible(True)
```
